```python
import json
from collections import Counter

# Load COCO JSON
with open("mock_coco.json", "r") as f:
    coco = json.load(f)

images = coco["images"]
annotations = coco["annotations"]
categories = coco.get("categories", [])

# Map category_id â†’ category_name
category_id_to_name = {cat["id"]: cat["name"] for cat in categories}

# Summary
total_images = len(images)
total_annotations = len(annotations)

# Count annotations per category
category_counts = Counter(a["category_id"] for a in annotations)

# Top 3 categories
top3 = category_counts.most_common(3)

print(f"Total Images: {total_images}")
print(f"Total Annotations: {total_annotations}")
print("Top 3 Categories:")
for cat_id, count in top3:
    name = category_id_to_name.get(cat_id, f"ID {cat_id}")
    print(f"  {name}: {count} annotations")
```

```python
from collections import Counter

def get_category_counts(image_ids, annotations, category_map):
    counts = Counter()
    for ann in annotations:
        if ann['image_id'] in image_ids:
            counts[category_map[ann['category_id']]] += 1
    return counts

category_id_to_name = {cat['id']: cat['name'] for cat in coco['categories']}

train_counts = get_category_counts(train_ids, coco['annotations'], category_id_to_name)
val_counts = get_category_counts(val_ids, coco['annotations'], category_id_to_name)
test_counts = get_category_counts(test_ids, coco['annotations'], category_id_to_name)

print("Train category distribution:", train_counts)
print("Validation category distribution:", val_counts)
print("Test category distribution:", test_counts)
```

```python
from collections import defaultdict

seen = set()
duplicates = []

for ann in coco['annotations']:
    key = (ann['image_id'], ann['category_id'], tuple(ann['bbox']))
    if key in seen:
        duplicates.append(ann['id'])
    else:
        seen.add(key)

print("Duplicate annotation IDs:", duplicates)
```
